```python
#Taken from - https://stackoverflow.com/questions/58254609/python-docx-parse-a-table-to-panda-dataframe

#Importing libraries
from docx import Document
import pandas as pd
import os
k = 0
table_list = {}
for q, filename in enumerate(os.listdir(r"C:\Users\dgthomas\Desktop\2018 - word")):
    #print(q)
    #Enter word document name here (with the directory)
    document = Document(r"C:\Users\dgthomas\Desktop\2018 - word" + "\\" + filename) 

    # Initializaing a list - 'table_list'
    #table_list = [[0 for x in range(100)] for y in range(300)]
    
    #Looping over all document tables and scrapping data
    for w, table in enumerate(document.tables):
        #print(w)
        df = [['' for i in range(len(table.columns))] for j in range(len(table.rows))]
        for i, row in enumerate(table.rows):
            for j, cell in enumerate(row.cells):
                if cell.text:
                    df[i][j] = cell.text
        if table_list.get(q) is None:
            table_list[q] = {}
            #print(q,"foo")
        
        table_list[q][w] = pd.DataFrame(df)
        #print("w = ",w,"table = ",table_list[0])
        #print(table_list[0] == None)
        #print('q = ',q,'w = ',w)
        #print(table_list[q][w])
            

    # Appending the table list
        #table_list[table].append(pd.DataFrame(df))
    #print(table_list[k])
    #k = k+1
    #del df[:]
```
