```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fields = ['Date of Crash', 'Time', 'Day of Week', 'County', 'City', 'Drug or Alcohol Related', 'Driver Age','Driver Gender','Driver Condition']
df = pd.read_csv('crash data june 19 2020 (1).csv',usecols = fields )
df.head()
```

```python
df1 = df[(df['Drug or Alcohol Related'] == 'Alcohol (Statutory)') | (df['Drug or Alcohol Related'] == 'Under Influence of Alcohol/Drugs/Medications')]
```

```python
df1.head()
```

```python
df1['Standardized_Dates'] = pd.to_datetime(df1['Date of Crash'], format='%Y%m%d')
df1.head()
```

```python
df2 = pd.read_excel('co-est2019-annres-19 (1).xlsx', header = None, usecols= [0,12], thousands=',')
```

```python
df2.head(10)
```

```python
df2 = df2.iloc[5:]
```

```python
df2.tail(-10)
```

```python
df2 = df2[:-5]
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df2. iloc[:, 0] = df2. iloc[:, 0].str.replace('County, Iowa', "",regex=True)
```

```python
df2. iloc[:, 0] = df2. iloc[:, 0].str.replace('.', "",regex=True)
```

```python
df2
```

```python
df2. iloc[:, 0] = df2. iloc[:, 0].str.replace('  *', "",regex=True)
```

```python
df2.columns = ['County', 'Population']
```

```python
df1
```

```python
df3=df1.merge(df2, on = ('County'))
```

```python
df3
```

```python
df4 = df3['County'].value_counts()
```

```python
df4 = pd.DataFrame(df4)
```

```python
df4.reset_index(level=0, inplace=True)
```

```python
df4.columns = ['County','Accident Counts']
```

```python
df4
```

```python
df5=df4.merge(df3, on = ('County'))
```

```python
df5['Accidents-Per-Capita'] = df5['Accident Counts']/df5['Population']
```

```python
df5
```

```python
import bokeh
from bokeh.plotting import figure, output_file, output_notebook,show
from bokeh.models import ColumnDataSource, CDSView, GroupFilter
from bokeh.models import ColumnDataSource, FactorRange
output_notebook()
```

```python
df6 = df5[['County', 'Accidents-Per-Capita']].copy()
```

```python
df6 = df6.drop_duplicates()
df6.sort_values('Accidents-Per-Capita', inplace=True)
```

```python
import math
source_Q6 = ColumnDataSource(df6)
p2 = figure(x_range=FactorRange(factors=df6['County']),plot_width=3000, plot_height=1000,title="Accidents per Capita per County")
p2.vbar(x = 'County', top = 'Accidents-Per-Capita', fill_alpha = 0.2, width=0.8,source = source_Q6 )
p2.xaxis.major_label_orientation = math.pi/2

show(p2)
```

```python
df7 = df5[['County', 'Accident Counts']].copy()
df7 = df7.drop_duplicates()
df7.sort_values('Accident Counts', inplace=True)
```

```python
df7.sort_values('Accident Counts', inplace=True)
```

```python
import math
source_Q5 = ColumnDataSource(df7)
p3 = figure(x_range=FactorRange(factors=df7['County']),plot_width=3000, plot_height=1000,title="Accidents per County")
p3.vbar(x = 'County', top = 'Accident Counts', fill_alpha = 0.2, width=0.8,source = source_Q5 )
p3.xaxis.major_label_orientation = math.pi/2
show(p3)
```
